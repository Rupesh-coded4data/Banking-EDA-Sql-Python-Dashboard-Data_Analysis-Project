```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine
connection_string ='mysql+pymysql://root:2103@localhost:3306/data_frame_db'
engine = create_engine(connection_string)
```

```python
query= "Select * From banking_case.customer"
```

```python
df = pd.read_sql(query, engine)
```

```python
bins = [0,100000,300000, float('inf')]
labels = ['Low','Medium','High']

df['income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)
```

```python
df['income Band'].value_counts().plot(kind='bar')
```

```python
#Examine the Distribution of Unique Catogories in Catogorical Columns
categorical_cols = df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation",
                      "Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","income Band"]].columns

for col in categorical_cols:
    print(f"Value Counts for '{col}':")
    display(df[col].value_counts())
    print("\n" + "-"*50 + "\n")
```

```python
#Univariate Analysis
```

```python
# Use a vibrant color palette for variety
palette = sns.color_palette("husl")  # Try "Set2", "pastel", or "deep" as alternatives

for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation",
                                  "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting",
                                  "income Band"]].columns):
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=predictor, palette=palette)
    plt.title(f"Distribution of {predictor}", fontsize=14, fontweight='bold')
    plt.xlabel(predictor, fontsize=12)
    plt.ylabel("Count", fontsize=12)
    plt.xticks(rotation=45, fontsize=10)
    plt.yticks(fontsize=10)
    sns.despine()  # Clean style: removes top and right border
    plt.tight_layout()
    plt.show()
```

#Bivariate Analysis

```python
palette = sns.color_palette("Set2")  # Try "husl", "pastel", or "deep" if you want variety
for i, predictor in enumerate(df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation",
                                  "Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting",
                                  "income Band"]].columns):
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=predictor, hue='GenderId', palette=palette)
    plt.title(f"Count of {predictor} by Gender", fontsize=14)
    plt.xlabel(predictor, fontsize=12)
    plt.ylabel("Count", fontsize=12)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
#Histplot of Values Counts for Different Occupation

# Set a clean style
sns.set_style('whitegrid')

for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(8, 4))
    
    # Use a colorful palette
    color = sns.color_palette("Set2", len(df[col].unique()))[0]
    sns.histplot(df[col], color=color)
    
    # Use the actual column name in the title
    plt.title(f"Histogram of {col} Count", fontsize=14, fontweight='bold', color='darkblue')
    plt.xlabel(col, fontsize=12, color='darkgreen')
    plt.ylabel("Count", fontsize=12, color='darkred')
    
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
```

#-----------------Numerical Analysis-----------------#

```python
numerical_cols = [
    'Estimated Income', 'Superannuation Savings', 'Credit Card Balance',
    'Bank Loans', 'Checking Accounts', 'Saving Accounts',
    'Foreign Currency Account', 'Business Lending'
]

# Set Seaborn theme
sns.set_style("whitegrid")

# Create 8 subplots (one for each column)
fig, axes = plt.subplots(nrows=8, ncols=1, figsize=(10, 24))  # Adjusted height

# Define 8 different colors
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728',
          '#9467bd', '#8c564b', '#e377c2', '#7f7f7f']

# Plot each numerical column
for i, col in enumerate(numerical_cols):
    sns.histplot(data=df, x=col, kde=True, ax=axes[i], color=colors[i])
    axes[i].set_title(f"Distribution of {col}", fontsize=14)
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
numerical_cols = [
    'Estimated Income', 'Superannuation Savings', 'Credit Card Balance',
    'Bank Loans', 'Checking Accounts', 'Saving Accounts',
    'Foreign Currency Account', 'Business Lending'
]

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    square=True,
    cbar_kws={'shrink': 0.8}
)
plt.title("Correlation Heatmap", fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
